## 1. How is volume of orders broken down by hour within a daily market?

I am going to take a look at order flow, first I need to figure out the min and max date for the dataset for the user trading analysis to make sure it is accurate as the data is from one NBA daily market.

```python
print(df_all['order_time'].min())
print(df_all['order_time'].max())
```

```python
df_all_live = df_all[df_all['phase'] == 'LIVE']
hourly_count = df_all.groupby(df_all['order_time'].dt.hour).sum().reset_index()
hourly_count_live = df_all_live.groupby(df_all['order_time'].dt.hour).sum().reset_index()
```

```python
for x in range(8):
    if hourly_count['order_time'][x] != 0:
        hourly_count['order_time'][x] = hourly_count['order_time'][x] - 12
    else:
        hourly_count['order_time'][x] = 12

        
hourly_count.sort_values('order_time', inplace=True)
hourly_count
```

```python
for x in range(6):
    if hourly_count_live['order_time'][x] != 0:
        hourly_count_live['order_time'][x] = hourly_count_live['order_time'][x] - 12
    else:
        hourly_count_live['order_time'][x] = 12

hourly_count_live.sort_values('order_time', inplace=True)
hourly_count_live
```

```python
fig, ax = plt.subplots(figsize = (10,8))
plt.xlabel('Market Hours (PM-AM)')
x = hourly_count.order_time
plt.xticks(x)
plt.ylabel('Number of Shares')
plt.title('Quantity of Shares Requested by Hour')
ax.plot(hourly_count.order_time, hourly_count.quantity)
ax.plot(hourly_count_live.order_time, hourly_count_live.quantity)
ax.legend(['All Orders', 'Live Orders'])
```

## 2. Count of Total Orders broken down by projected rank

Since there are numerous datapoints that fall within the orders and projected rank columns and it would be difficult to visualize all within one chart, I will first bucket the orders and projected rank columns and then create a count plot in seaborn to visualize the data.

```python
labels = ['rank_1_10', 'rank_11_20', 'rank_21_30', 'rank_31_40', 'rank_41_50', 'rank_51_60', 'rank_61_70', 'rank_71_80']
buckets = [0, 10, 20, 30, 40, 50, 60, 70, 81]
df_all['projected_ranked_buckets'] = pd.cut(x= df_all['rank_projected'], bins=buckets, labels=labels)
```

```python
labels = ['quantity_1_10', 'quantity_11_20', 'quantity_21_30', 'quantity_31_40', 'quantit|y_41_50', 'quantity_51_60', 'quantity_61_70', 'quantity_71_82']
buckets = [0, 10, 20, 30, 40, 50, 60, 70, 83]
df_all['order_quantity_buckets'] = pd.cut(x= df_all['quantity'], bins=buckets, labels=labels)
```

```python
sns.catplot(x='projected_ranked_buckets', hue='order_quantity_buckets',kind='count', data=df_all, 
            height=8, aspect=1, legend_out=False, dodge=False)
plt.xlabel('Projected Ranking Buckets')
plt.ylabel('Count of Orders')
plt.title('Count of Orders by Projected Rank')
plt.show()
```

## 4. Are shares bought during IPO more likely to be profitable vs shares bought live?

For this, I will need to create a new "profit_loss" column using the available data for each order and then chart whether an order is more likely to be profitable if it is placed during the IPO phase or the Live phase.

```python
df_all['profit_loss'] = df_all['payout_final'] * df_all['filled_quantity'] - df_all['cost_basis']
df_all
```

```python
df_all['profitable'] = np.where(df_all['profit_loss'] > 0, 'yes', 'no')
df_all
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
perc_prof_live = round(df_all[(df_all.profit_loss >= 0) & (df_all.side == 'BUY') &(df_all.phase == 'LIVE')].shape[0] / df_all[df_all.phase == 'LIVE'].shape[0], 2)
perc_prof_ipo = round(df_all[(df_all.profit_loss >= 0) & (df_all.side == 'BUY') &(df_all.phase == 'IPO')].shape[0] / df_all[df_all.phase == 'IPO'].shape[0], 2)

fig, ax = plt.subplots(figsize =(12, 8))
plt.bar(['Live', 'IPO'], [perc_prof_live, perc_prof_ipo], color='#5975a4', width=0.5)
plt.xlabel('Phase')
plt.ylabel('Proportion of Profitable Orders')
plt.title('Proportion of Profitable Orders by Phase')
addlabels(['Live', 'IPO'],[perc_prof_live, perc_prof_ipo])
```
